```python
import pandas as pd
```

```python
data=pd.read_excel('housing_dataset.xlsx')
```

```python
data
```

```python
null_column=data.isna().sum()
```

```python
data['Price'].fillna(data['Price'].median())
```

```python
data['Bathrooms'].fillna(data['Bathrooms'].mean())
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
q1=data['Price'].quantile(0.25)
q1
```

```python
q3=data['Price'].quantile(0.75)
q3
```

```python
IQR=q3-q1
```

```python
lb=q1-1.5*IQR
```

```python
ub=q1+1.5*IQR
```

```python
data
```

```python
outliers=data[(data['Price']<lb) | (data['Price']>ub)]
```

```python
outliers
```

```python
cleaned=data[(data['Price']>=lb) & (data['Price']<=ub)]
```

```python
cleaned
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
Scaler=MinMaxScaler()
```

```python
cleaned['Sqft_Living']=Scaler.fit_transform(cleaned[['Sqft_Living']])
```
